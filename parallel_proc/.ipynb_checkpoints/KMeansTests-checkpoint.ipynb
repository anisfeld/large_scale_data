# Parallelizing K-means

In this notebook, I run tests on three implementations of K-means clustering. First we use no parallel processing. Second, we use a multiprocessing queue and specifically assign chunks of data to the processors. Third, we use a multiprocessing pool which automates the assignment to different processors.

```python
from KMeans import *
import pandas as pd
import numpy as np

data = pd.read_csv("County_Mortgage_Funding.csv")
data = data.drop(data.columns[0], axis=1)
d = data.iloc[:,[3,11]]
dcols = d.columns
```

# Baselines 

Before testing the different implementations on the full dataset, we start with a 2 feature dataset. This allows the clustering to be easily visualized. We see that each implementation returns the same centroid.

```python
centers, assigns = k_means(d,3, random=False)
df2 = pd.DataFrame(np.c_[assigns, d.copy()], columns=['group', dcols[0], dcols[1]])
kmeans2way(df2)
```

```python
centers, assigns = parallel_proc_k_means(d,3, random=False)
df2 = pd.DataFrame(np.c_[assigns, d.copy()], columns=['group', dcols[0], dcols[1]])
kmeans2way(df2)
```

```python
centers, assigns = pool_k_means(d,3, random=False)
t = np.c_[assigns, d.copy()]
df2 = pd.DataFrame(t, columns=['group', dcols[0], dcols[1]])
kmeans2way(df2)
```

# number of procs

The number of processors is not directy proportional to efficiency gains. Rather, there's a tradeoff between cost of engaging different processors and the gain from doing so. 

```python
for nprocs in [1,2,3,4]:
    print("parallel_proc with 3 means and {} cores".format(nprocs))
    %timeit parallel_proc_k_means(data, 3, nprocs=nprocs, TIMER_2=False)
```

```python
for nprocs in [1,2,3,4]:
    print("pool with 3 means and {} cores".format(nprocs))
    %timeit pool_k_means(data, 3, nprocs=nprocs, TIMER_2=False)
```

# Adding more means

```python
for n in range(1,10):
    print("parallel_proc with {} means and 3 cores".format(n))
    %timeit parallel_proc_k_means(data, n, TIMER_2=False)
```

```python
for n in range(1,10):
    print("pool with {} means and 3 cores".format(n))
    %timeit pool_k_means(data, n,  TIMER_2=False)
```


## Chunksize with mp.Pool.imap

The amount of data you send to a worker has a big effect on efficiency. The mp.Pool.imap function allows us to set the chunksize, which is the amount of data sent to each processor. The default value is 1, which performs very poorly. The algorithm takes roughly 10 times longer per iteration using chunk size 1 compared to chunk size 3000.

```python
for chunk in [1,30,300,3000, 10458]:
    print("Chunksize = {}".format(chunk))
    %timeit pool_k_means(data, 3, chunksize=chunk, TIMER=False, TIMER_2=False)
```

## distance functions 

np.linalg.norm is less efficient than manually calculating the L2 norm. This likely reflects the overhead builtin to np.linalg.norm to avoid errors/bugs.

```python
v1 = (np.array(range(12*3)) - 34.320).reshape(3, 12)
v2 = np.array(range(12))

def distance_lin(v1, v2):
    if len(v1.shape)==1: 
        v1 = v1[np.newaxis,:]; 
    return np.linalg.norm(v2 - v1, axis=1)

def distance(v1, v2):
    if len(v1.shape)==1:
        v1 = v1[np.newaxis,:]
    return np.sqrt(np.sum((v1 - v2)**2, axis=1))
    
```

```python
%timeit distance(v1, v2)
%timeit distance_lin(v1, v2)
```
